```python
# Now we can import yfinance (you may need to restart the kernel)

import yfinance as yf
import pandas as pd
from pathlib import Path
```

## You need to consult `pandas` documentation https://pandas.pydata.org/pandas-docs/stable/user_guide/10min.html to complete this assignment

```python
# It's practical to download data from Yahoo!Finance just once and keep a local copy
if Path(DATAFILE_NAME).is_file(): # check if we got the datafile already
    data = pd.read_pickle(DATAFILE_NAME) # read it
else:
    data = yf.download(tickers=TICKERS, start=START, end=END, group_by='ticker', auto_adjust=True)
    data.to_pickle(DATAFILE_NAME) # save localy
```

Now that you've read the Pandas documentation and have learned how to manipulate financial data, your task is to analyze the 84 downloaded stock datasets. Specifically, identify the stock that experienced the largest single-day High-Low price difference.

- What is the name of the company?
- On what date did this occur?
- What was the percentage difference of this daily High-Low fluctuation?

```python
# Your code goes here.
#Make a new dataframe with just the differences between 'High' and 'Low' values.
diff_data = (data.xs('High', level=1, axis=1) - data.xs('Low', level=1, axis=1))
```

```python
#Now we can find the maximum difference between the 'High' and 'Low' values, 
#and the company which had this difference and the day when this occured.
#diff_data.max().idxmax(), diff_data[diff_data.max().idxmax()].idxmax(), diff_data.max().max()
print(f"Company name: {diff_data.max().idxmax()}")
print(f"Date: {diff_data[diff_data.max().idxmax()].idxmax()}")
print(f"Largest single-day High-Low price difference: {diff_data.max().max()}")
```

```python
#Verify the Date and the Hig-Low price Difference for FICO company manually
maxday_date = pd.to_datetime('2022-01-28')
data_FICO = data.loc[maxday_date,('FICO',['High','Low'])]
print(data_FICO)
print('Difference',497.0-425.72)
```

```python
#Calculate the percentage difference of daily High-low fluctuation
percentagediff_data = ( (data.xs('High', level=1, axis=1) - data.xs('Low', level=1, axis=1) ) / data.xs('Low', level=1, axis=1) )*100
print(percentagediff_data.head())
```

```python
#Locate the pecentage corresponding to largest single day Hig-Low price difference
percentage_FICO = percentagediff_data.loc[maxday_date,('FICO')]
print(f"Percentage: {percentage_FICO}%")
```
